Euclidean distance__use modularity to run clustering 

```python
import numpy as np
import pandas as pd
import scipy.io
import community
import networkx as nx
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans
from sklearn import metrics

XYZ_Coord = scipy.io.loadmat('XYZcoord1.mat')
XYZ_C = np.array(XYZ_Coord ['XYZ1'])
print(XYZ_C)

XYZ_Edges = pd.read_csv('Edges.csv')
print(XYZ_Edges)


z = XYZ_C[:,2]
x = XYZ_C[:,0]
y = XYZ_C[:,1]
XYZ_E = np.zeros((XYZ_Edges.shape[0],3))
source = XYZ_Edges['Source'].to_numpy()
target = XYZ_Edges['Target'].to_numpy()
for i in range(source.shape[0]):
    distance = np.sqrt((x[target[i]-1] - x[source[i]]-1)**2 + (y[target[i]-1] - y[source[i]])**2 + (z[target[i]-1] - z[source[i]-1])**2)
    XYZ_E[i][0] = source[i]
    XYZ_E[i][1] = target[i]
    XYZ_E[i][2] = distance
print(XYZ_E)
print()

#Use pre-defined linkage (Edges.csv) to constructure whole network
G = nx.Graph()
for i in range(0, len(XYZ_E)):    
    e = ( str(int(XYZ_E[i,0])), str(int(XYZ_E[i,1])), XYZ_E[i,2] )
    G.add_weighted_edges_from([(e)])  
    
partition = community.best_partition(G)
size = float(len(set(partition.values())))
print("community:", size)
mod = community.modularity(partition,G)
print("modularity:", mod)

```

```python
#assign node color based on community in network
for com in set(partition.values()) :
    #print("Community", com)
    members = list_nodes = [nodes for nodes in partition.keys() if partition[nodes] == com]
    #print(members)
values = [partition.get(node) for node in G.nodes()]
nx.draw_spring(G, cmap = plt.get_cmap('jet'), node_color = values, node_size = 30, with_labels = False)
plt.title('Graph__euclidean distance')
plt.show()    

#assign data point color based on community in realspace
label = np.zeros((len(XYZ_C),1))
for j in set(partition.values()) :
    for i in range(len(XYZ_C)) :
        if partition[str(i+1)] == j :            
            label[i] =  j
labelRE = np.reshape(label, len(XYZ_C))            

```

```python
plt.figure(1)
ax = plt.axes(projection='3d')
z = XYZ_C[:,2]
x = XYZ_C[:,0]
y = XYZ_C[:,1]
c = labelRE
ax.scatter(x, y, z, c = c, cmap = plt.get_cmap('jet'))
ax.view_init(80, 0)
plt.title('Cluster result 1 by modularity')


plt.figure(2)
ax = plt.axes(projection='3d')
z = XYZ_C[:,2]
x = XYZ_C[:,0]
y = XYZ_C[:,1]
c = labelRE
ax.scatter(x, y, z, c = c, cmap = plt.get_cmap('jet'))
ax.view_init(45, 0)
plt.title('Cluster result 2 by modularity')
plt.show()
```

Euclidean distance__use  k-mean to run clustering 

```python
best = -2
for j in range(int(size)):
    kmeans = KMeans(n_clusters = (j+1), max_iter = 1000).fit(XYZ_E)
    label = [0] * XYZ_C.shape[0]
    for i in range(len(label)) :
        label[i] = np.zeros((1, int(size))).astype(int)
    for i in range(kmeans.labels_.shape[0]) :
        label[source[i]-1][0][kmeans.labels_[i]] = label[source[i]-1][0][kmeans.labels_[i]] + 1
        label[target[i]-1][0][kmeans.labels_[i]] = label[target[i]-1][0][kmeans.labels_[i]] + 1
    for i in range(len(XYZ_C)) :
        label[i] = np.argmax(label[i][0])
    #labelRE = np.reshape(label, len(XYZ_C))
    avg = metrics.silhouette_score(XYZ_C, labelRE)
    if avg > best :
        best = avg
        final_label = labelRE
final_label = np.reshape(final_label, len(XYZ_C))
```

```python

plt.figure(3)
ax = plt.axes(projection='3d')
z = XYZ_C[:,2]
x = XYZ_C[:,0]
y = XYZ_C[:,1]
c = final_label
ax.scatter(x, y, z, c = c, cmap = plt.get_cmap('jet'))
ax.view_init(80, 0)
plt.title('Cluster result 1 by Kmeans')

plt.figure(4)
ax = plt.axes(projection='3d')
z = XYZ_C[:,2]
x = XYZ_C[:,0]
y = XYZ_C[:,1]
c = final_label
ax.scatter(x, y, z, c = c, cmap = plt.get_cmap('jet'))
ax.view_init(45, 0)
plt.title('Cluster result 2 by Kmeans')
plt.show()
```
